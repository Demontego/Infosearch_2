# Analysis/Modeling
Do work here

```python
sites=set()
graph_1={}
graph_2={}
h_a={}
with open("link_graph_series.txt", 'r', encoding='utf-8') as f:
    for line in f:
        x,y= line.strip().split('\t')
        sites.add(x)
        sites.add(y)
        if x in graph_1:
            graph_1[x].append(y)
        else:
            graph_1[x]= []
            graph_1[x].append(y)
        if y in graph_2:
            graph_2[y].append(x)
        else:
            graph_2[y]= []
            graph_2[y].append(x)
        if x not in h_a:
            h_a[x]=(1e-20,1e-20)
        if y not in h_a:
            h_a[y]=(1e-20,1e-20)
```

```python
import copy
import operator
```

```python
for i in range(20):
    tmp={}
    for k,site in enumerate(sites):
        tmp_h=0.00001
        tmp_a=0.00001
        if site in graph_1:
            tmp_h=np.sum([h_a[sit][1] for sit in graph_1[site]])
        if site in graph_2:   
            tmp_a=np.sum([h_a[sit][0] for sit in graph_2[site]])
        tmp[site]=(tmp_h,tmp_a)
        if k%100000==0:
            print(site, tmp[site])
    h_a=copy.deepcopy(tmp)
```

```python
hs_={}
as_={}
for i in h_a.keys():
    hs_[i]=h_a[i][0]
    as_[i]=h_a[i][1]
```

```python
a_=sorted(as_.items(), key=operator.itemgetter(1), reverse=True)
h_=sorted(hs_.items(), key=operator.itemgetter(1), reverse=True)
```

```python
for i,k in a_:
    if 'person/' in i:
        print(i,k)
        break
```

```python
for i,k in h_:
    if 'series_' in i:
        print(i,k)
        break
```
